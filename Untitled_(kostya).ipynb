# Общая информация

```python
import pandas as pd

#Загрузка файлов
ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
tags = pd.read_csv("tags.csv")
genome_tags = pd.read_csv("genome-tags.csv")
genome_scores = pd.read_csv("genome-scores.csv")

#Прочитаем информацию по файлам
eda_summary = {
    "ratings": ratings.info(),
    "movies": movies.info(),
    "tags": tags.info(),
    "genome_tags": genome_tags.info(),
    "genome_scores": genome_scores.info()
}
```

# Распределения

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter

#Распредления оценок
rating_distribution = ratings['rating'].value_counts().sort_index()

#Топ-10 популярных по количеству рейтинга фильмов
top_movies_by_rating_count = ratings['movieId'].value_counts().head(10)
top_movies_df = movies[movies['movieId'].isin(top_movies_by_rating_count.index)]
top_movies_df['rating_count'] = top_movies_df['movieId'].map(top_movies_by_rating_count)

#Популярные жанры
all_genres = movies['genres'].str.split('|').sum()
genre_counts = pd.Series(Counter(all_genres)).sort_values(ascending=False)

#Активные пользователи
top_users = ratings['userId'].value_counts().head(10)

#Подсчёт матрицы плотности
num_users = ratings['userId'].nunique()
num_items = ratings['movieId'].nunique()
matrix_density = len(ratings) / (num_users * num_items)

#Вывод метрик
eda_stats = pd.DataFrame({
    "Метрика": [
        "Всего оценок",
        "Уникальных пользователей",
        "Уникальных фильмов",
        "Плотность матрицы (чем ближе к 1 — тем плотнее)",
        "Уникальных жанров"
    ],
    "Значение": [
        len(ratings),
        num_users,
        num_items,
        round(matrix_density, 5),
        len(genre_counts)
    ]
})

print(eda_stats)

# Графики
plt.figure(figsize=(10, 4))
sns.barplot(x=rating_distribution.index, y=rating_distribution.values, palette="viridis")
plt.title("Распределение оценок")
plt.xlabel("Оценка")
plt.ylabel("Количество")
plt.show()

plt.figure(figsize=(10, 4))
sns.barplot(x=top_movies_df['title'], y=top_movies_df['rating_count'], palette="magma")
plt.title("Топ-10 фильмов по количеству оценок")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Кол-во оценок")
plt.xlabel("Фильм")
plt.show()

plt.figure(figsize=(10, 4))
sns.barplot(x=genre_counts.head(10).index, y=genre_counts.head(10).values, palette="coolwarm")
plt.title("Топ-10 жанров")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Частота")
plt.xlabel("Жанр")
plt.show()

plt.figure(figsize=(10, 4))
sns.barplot(x=top_users.index.astype(str), y=top_users.values, palette="cubehelix")
plt.title("Топ-10 самых активных пользователей")
plt.xlabel("userId")
plt.ylabel("Кол-во оценок")
plt.show()
```

# Сравнение пользователей между собой (user-based) - Черновик

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix


ratings = pd.read_csv("ratings.csv")

#Отбор 1000 активных пользователей по числу оценок для облегчения и ускорения анализа
top_users = ratings['userId'].value_counts().head(1000).index
ratings_top_users = ratings[ratings['userId'].isin(top_users)]

#Отбор 1000 просматриваемых фильмов по числу оценок для облегчения и ускорения анализа
top_movies = ratings_top_users['movieId'].value_counts().head(1000).index
ratings_filtered = ratings_top_users[ratings_top_users['movieId'].isin(top_movies)]

#Построение user-item матрицы
user_item_matrix = ratings_filtered.pivot(index='userId', columns='movieId', values='rating').fillna(0)

user_item_df = user_item_matrix.reset_index()

sparse_matrix = csr_matrix(user_item_matrix.values)

user_ids = user_item_matrix.values

similarity_matrix = cosine_similarity(sparse_matrix)


similarity_df = pd.DataFrame(similarity_matrix, index=user_item_matrix.index, columns=user_item_matrix.index)


top_similar_users = {}

for user_id in similarity_df.index:
    similar_users = similarity_df.loc[user_id].drop(user_id).sort_values(ascending=False).head(5)
    top_similar_users[user_id] = list(similar_users.index)


# Пример вывода
for uid, sim_users in list(top_similar_users.items())[:5]:
    print(f"userId {uid} → похожие: {sim_users}")
```

### t-SNE пользователей

```python
from sklearn.manifold import TSNE

# Применим t-SNE к матрице user-item
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
user_embeddings = tsne.fit_transform(sparse_matrix.toarray())

# Визуализируем
plt.figure(figsize=(10, 8))
plt.scatter(user_embeddings[:, 0], user_embeddings[:, 1], s=10, alpha=0.7)
plt.title("Карта пользователей (t-SNE)")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

### Кластеризация (kMeans) для сегментации пользователей

Воспользуемся методом локтя, чтобы определить оптимальное число кластеров.

```python
from sklearn.cluster import KMeans
import numpy as np
inertia = []
K_range = range(2, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(user_embeddings)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o')
plt.xlabel('Количество кластеров')
plt.ylabel('Инерция')
plt.title('Метод локтя для выбора k')
plt.show()
```

Видим, что после 6 кластеров шаги инерции уже минимальны. Остаёмся на этом значении.

```python
from sklearn.cluster import KMeans

#Кластеризация
k = 6

kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(user_embeddings)
```

Присвоим нашим 1000 активным юзерам номера кластеров для дальнейшего анализа.

```python
user_ids = user_item_df['userId']
cluster_df = pd.DataFrame({
    'userId': user_ids,
    'cluster': clusters
})
```

### Анализ кластеров

Для исследования кластеров, можем привести базовый анализ: посчитать количество пользователей, популярные просматриваемые фильмы и жанры.

```python
#Мерджим таблицы рейтинга и кластеров
merged = ratings.merge(cluster_df,on='userId')

#Посчитаем, сколько юзеров у нас в каждом кластере
cluster_sizes = cluster_df['cluster'].value_counts().sort_index()
print(cluster_sizes)

#Сделаем подсчёт количества и сумму оценок, а также средную оценку фильма у пользователей
cluster_top_movies = (
    merged.groupby(['cluster', 'movieId'])['rating']
    .agg(['count','sum'])
    .reset_index()
    .sort_values(['cluster', 'count','sum'], ascending=False)
)

cluster_top_movies['mean'] = cluster_top_movies['sum']/cluster_top_movies['count']

# Выведем топ-10 фильмов по каждому кластеру
for i in range(cluster_df['cluster'].nunique()):
    print(f"Кластер {i}")
    top5 = cluster_top_movies[cluster_top_movies['cluster'] == i].head(10)
    display(top5.merge(movies, on='movieId')[['title', 'count','sum','mean']])
```

```python
#По такой же схеме теперь проанализируем Жанры у пользователей
merged_genre = merged.merge(movies,on='movieId')
merged_genre
```

```python
#Создаём список жанров (отделяем их и "взрываем", те трансформиуем лист в строки)
merged_genre['genre_list'] = merged_genre['genres'].str.split('|')
merged_genre_exploded = merged_genre.explode('genre_list')

#считаем количество втсречающихся жанров по кластерам
genre_counts = (
    merged_genre_exploded.groupby(['cluster','genre_list'])
    .size()
    .reset_index(name='count')
    .sort_values(['cluster','count'],ascending=False)
)

# Выведем топ-10 фильмов по каждому кластеру
for i in range(cluster_df['cluster'].nunique()):
    print(f"Кластер {i}")
    top_genre = genre_counts[genre_counts['cluster'] == i].head(10)
    display(top_genre[['genre_list', 'count']])
```

Выводы:

*   провели анализ данных по видеохостингу и построили распределения для будущей работы над рекомендательными системами;
*   Воспользовались t-SNE и kMeans для визуализации и анализа групп пользователей с похожими вкусами для создания рекомендательных систем (рекомендации на основе популярных фильмов, любимых жанров). Для облечгения сделали выборку из топ-1000, в дальнейшем можно масштабировать на весь сет.

```python
tags = pd.read_csv("tags.csv")
genome_tags = pd.read_csv("genome-tags.csv")
genome_scores = pd.read_csv("genome-scores.csv")

tags_df = tags.merge(genome_tags, on='tag')
tags_df
```

```python
tags_count = tags_df.groupby(['movieId','tagId','tag'])['tag'].count().reset_index(name='count').sort_values(['count'],ascending=False)
tags_count
```

```python
from collections import Counter

tags_list = tags_df['tag']
tags_counts = pd.Series(Counter(tags_list)).sort_values(ascending=False)

plt.figure(figsize=(10, 4))
sns.barplot(x=tags_counts.head(10).index, y=tags_counts.head(10).values, palette="coolwarm")
plt.title("Топ-10 Тегов")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Количество")
plt.xlabel("Тег")
plt.show()
```

```python
tags_frequency = genome_scores.groupby('tagId')['relevance'].sum().reset_index().sort_values(by='relevance',ascending=False)
tags_frequency = tags_frequency.merge(genome_tags,on='tagId')

plt.figure(figsize=(10, 4))
sns.barplot(x=tags_frequency['relevance'].head(20), y=tags_frequency['tag'].head(20).values, palette="coolwarm")
plt.title("Топ-20 Тегов по суммарной релевантности")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Тег")
plt.xlabel("Сумма релевантности (веса)")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(genome_scores['relevance'],bins=50,kde=True)
plt.title('Распределение релевантности тегов по фильмам')
plt.xlabel('Средняя релевантность')
plt.ylabel('Частота фильмов')
plt.show()
```

Большинство тегов имеют низкую релевантность, есть явно значимые, но их количество очень мало.

```python
avr_tags_per_movie = genome_scores.groupby('movieId')['relevance'].mean()
sns.histplot(avr_tags_per_movie,bins=50,kde=True)
plt.title('Распределение средней релевантности тегов на фильм')
plt.xlabel('Средняя релевантность')
plt.ylabel('Частота фильмов')
plt.show()
```
